```python
def num_of_nulls(row):
    return row.isnull().sum()

null_values = pd.DataFrame(data.apply(num_of_nulls, axis=1), columns=["Nulls"])
null_values["Year"] = data["Year"]
null_values
```

```python
plt.figure(figsize=(10, 10))
plt.xlim([1500, 2000])
sns.scatterplot(x=null_values["Year"], y=null_values["Nulls"])
```

```python
mean_nulls_by_year = null_values["Nulls"].groupby(null_values["Year"]).mean()
```

```python
sns.scatterplot(x=mean_nulls_by_year.index, y=mean_nulls_by_year)
```

```python
year_counts = data.groupby(data["Year"]).apply(lambda rows: len(rows))
year_counts = year_counts.reset_index()
year_counts = year_counts.rename({0: "Count"}, axis=1)
year_counts
```

```python
year_counts = year_counts[year_counts["Year"] >= 1950]
year_counts["Year"] = (year_counts["Year"] % 100)
plt.figure(figsize=(18, 10))
sns.barplot(year_counts, x="Year", y="Count")
```

```python
nulls_by_year =  data.isnull()
nulls_by_year["Year"] = data["Year"]
nulls_by_year = nulls_by_year.groupby(nulls_by_year["Year"]).sum() / nulls_by_year.groupby(nulls_by_year["Year"]).count()
nulls_by_year
```

```python
condition = (nulls_by_year["Month"] == 0) & (nulls_by_year["Day"] == 0) & (nulls_by_year["EQ Primary"] <= 0.1) & (nulls_by_year["Coordinates"] == 0) & (nulls_by_year["Country"] == 0) & (nulls_by_year["Earthquake : Damage Description"] <= 0.3)
acceptable_years_nulls = nulls_by_year[condition]
acceptable_years_nulls
```
