```python
data1 = data
path_to_dataset2 = "significant-earthquake-database.csv"
data2 = pd.read_csv(path_to_dataset2, sep=";")
data2 = data2.sort_values(["Year", "Month", "Day", "EQ Primary"], ascending=True)
data2.reset_index(inplace=True, drop=True)
data2 = data2.add_suffix("_2")
data2[['Latitude_2','Longitude_2']] = data2['Coordinates_2'].str.split(',',expand=True)
data2["Latitude_2"] = data2["Latitude_2"].astype("float64")
data2["Longitude_2"] = data2["Longitude_2"].astype("float64")
data1 = data1.sort_values(["Year", "Mo", "Dy", "Mag"], ascending=True)
data1 = data1.add_suffix("_1")
data1.reset_index(inplace=True, drop=True)
data2.head(20)
```

```python
print(data1.shape)
print(data2.shape)
```

```python
data = pd.concat([data1, data2], axis=1)
print(data.shape)
data.head()
```

```python
for k, v in data.iloc[10].items():
    print(f"{k}: {v}")
```

```python
columns1 = ["Year_1", "Mo_1", "Dy_1", "Latitude_1", "Longitude_1", "Focal Depth (km)_1", "Mag_1", "MMI Int_1"]
columns2 = ["Year_2", "Month_2", "Day_2", "Latitude_2", "Longitude_2", "Focal Depth_2", "EQ Primary_2", "Intensity_2"]

def compare(row):
    same = True
    print(row.name)
    for col1, col2 in zip(columns1, columns2):
        if math.isnan(row[col1]) and math.isnan(row[col2]):
            continue
        if row[col1] != row[col2]:
            print(f"{col1}({type(row[col1])}): {row[col1]}, {col2}({type[row[col2]]}): {row[col2]}", end=" | ")
            same = False
    if(not same):
        print()
    return same

res = data.apply(compare, axis=1)
res
```

```python
from sklearn.metrics import jaccard_score
import pandas as pd

# Create a sample DataFrame with two columns of categorical data
df = pd.DataFrame({'Column1': ['A', 'B', 'C', 'D'], 'Column2': ['A', 'B', 'C', 'D']})

# Calculate the Jaccard similarity between the two columns
jaccard_similarity = jaccard_score(df['Column1'], df['Column2'], average='weighted')

print(jaccard_similarity)
```
